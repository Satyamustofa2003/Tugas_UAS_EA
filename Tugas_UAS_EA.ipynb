```
# Mengimpor pustaka yang diperlukan
import matplotlib.pyplot as plt

# Definisikan komponen arsitektur
servers = ['Server 1', 'Server 2', 'Server 3']
database = ['Database 1', 'Database 2']
applications = ['App 1', 'App 2', 'App 3']

# Definisikan diagram arsitektur
fig, ax = plt.subplots()
ax.set_title('Contoh Enterprise Architecture Design')

# Menampilkan server
for i, server in enumerate(servers):
    ax.text(0.1, 0.9 - i * 0.1, server, transform=ax.transAxes, bbox=dict(facecolor='red', alpha=0.5))

# Menampilkan database
for i, db in enumerate(database):
    ax.text(0.5, 0.9 - i * 0.1, db, transform=ax.transAxes, bbox=dict(facecolor='blue', alpha=0.5))

# Menampilkan aplikasi
for i, app in enumerate(applications):
    ax.text(0.9, 0.9 - i * 0.1, app, transform=ax.transAxes, bbox=dict(facecolor='green', alpha=0.5))

# Menampilkan diagram
plt.axis('off')
plt.show()
```

```
class Server:
    def __init__(self, name, ip_address, location):
        self.name = name
        self.ip_address = ip_address
        self.location = location

class Database:
    def __init__(self, name, type, location):
        self.name = name
        self.type = type
        self.location = location

class Application:
    def __init__(self, name, version, owner):
        self.name = name
        self.version = version
        self.owner = owner

# Contoh instansiasi beberapa komponen
server1 = Server("Server 1", "192.168.1.1", "Data Center A")
database1 = Database("Database 1", "SQL", "Data Center A")
application1 = Application("App 1", "v1.0", "John Doe")

# Menampilkan informasi komponen
print("Server Name:", server1.name)
print("Server IP Address:", server1.ip_address)
print("Server Location:", server1.location)

print("Database Name:", database1.name)
print("Database Type:", database1.type)
print("Database Location:", database1.location)

print("Application Name:", application1.name)
print("Application Version:", application1.version)
print("Application Owner:", application1.owner)
```
